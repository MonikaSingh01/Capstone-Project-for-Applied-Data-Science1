```python
import numpy as np
# TASK 3: Calculate the distances between a launch site to its proximities
# Example proximities DataFrame (replace with your actual proximities data)
# proximities_df should have columns: 'Proximity_Name', 'Lat', 'Long'
proximities_df = pd.DataFrame({
    'Proximity_Name': ['Location1', 'Location2'],
    'Lat': [28.5623, 34.6328],
    'Long': [-80.5774, -120.6108]
})

# Haversine function to calculate distance between two lat/lon points
def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arcsin(np.sqrt(a))
    R = 6371.0  # Earth radius in kilometers
    return R * c

# Calculate distances between each launch site and each proximity
distances = []
for i, launch_row in spacex_df.iterrows():
    for j, prox_row in proximities_df.iterrows():
        dist = haversine(launch_row['Lat'], launch_row['Long'], prox_row['Lat'], prox_row['Long'])
        distances.append({
            'Launch Site': launch_row['Launch Site'],
            'Proximity_Name': prox_row['Proximity_Name'],
            'Distance_km': dist
        })

distances_df = pd.DataFrame(distances)

# Show the first few rows
print(distances_df.head())
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
